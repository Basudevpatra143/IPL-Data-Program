```python
# 5. Create a story around year wise changes in the overall IPL performances.
# For that, get the records of year wise total run scored, total catches and total wickets taken, including all the player records. 
# Comment on the year-wise variations in them and support your comments using proper visualizations. 
```

```python
# Story:- every years win ipl in one team such us 2008 match win
# in rajasthan Royals and 2009 match win in Deccan chargers because
# ipl match shifted to South Africa and 2010 match win in csk and some year win in 
# one team. I am so all data:-
```

```python
# Calculate year-wise total runs scored, total catches taken, and total wickets taken

yearly_totals = df.groupby('Year').agg({
    'Runs_Scored': 'sum',
    'Catches_Taken': 'sum',
    'Wickets_Taken': 'sum'
}).reset_index()
yearly_totals
```

```python
# Plotting the year-wise total runs scored

plt.figure(figsize=(14, 8))
plt.plot(yearly_totals['Year'], yearly_totals['Runs_Scored'], marker='o', label='Total Runs Scored')
plt.title('Year-Wise Total Runs Scored in IPL')
plt.xlabel('Year')
plt.ylabel('Total Runs Scored')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting the year-wise total catches taken

plt.figure(figsize=(14, 8))
plt.plot(yearly_totals['Year'], yearly_totals['Catches_Taken'], marker='x', linestyle='--', label='Total Catches Taken')
plt.title('Year-Wise Total Catches Taken in IPL')
plt.xlabel('Year')
plt.ylabel('Total Catches Taken')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting the year-wise total wickets taken

plt.figure(figsize=(14, 8))
plt.plot(yearly_totals['Year'], yearly_totals['Wickets_Taken'], marker='s', linestyle='-.', label='Total Wickets Taken')
plt.title('Year-Wise Total Wickets Taken in IPL')
plt.xlabel('Year')
plt.ylabel('Total Wickets Taken')
plt.legend()
plt.grid(True)
plt.show()
```
